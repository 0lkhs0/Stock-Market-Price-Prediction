```python
# using yfinance which calls the yahoo finance api for daily stock and index prices
import yfinance as yf
import pandas as pd
```

```python
# SP500 = GSPC symbol 
sp500 = yf.Ticker("^GSPC")
```

```python
#query all data from the beginning
sp500 = sp500.history(period="max")
```

```python
sp500.index
```

```python
#observe the trend of SP500 from its beginning to now
sp500.plot.line(y="Close", use_index=True)
```

```python
# information is more useful for individual stocks not index fund
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500
```

```python
#setting up the column tomorrow so it can be used see if the price went up or down
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
# check if tomorrow's price is higher or lower than today's price
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
# going to far back isnt always good as the market could have shifted fundamentally
# .copy() to prevent pandas warning 

sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
# random forest classifier -> train a bunch of individual decision tress with randomised parameters then averaging them 
# resistent to overfitting 
# can pick up non-linear tendencies in the data

from sklearn.ensemble import RandomForestClassifier

# n_estimators -> number of individual decision trees to train (higher better)
# min_sample_split -> protect against overfitting  (higher less acc)
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
predictors = ["Close", "Volume", "Open", "High", "Low"]

model.fit(train[predictors], train["Target"])

```

```python
# use to measure how acc it really is
#percentage of the time when we said the market would go up did it actually go up
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
# a visualisation of when the model actually makes the correct decision
combined.plot()
```

```python
# function to fit the model and using the training predictors to generate predictions
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)

    return combined
```

```python
# 2500 gives us 10 years of trading data to train the first model
# step 250 means every 11th year use 10 years worth of data 12th use 11 years of data
# each finanical year has 250 days
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train,test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# give the model more information to work with
# rolling mean between the last 2 days / (5) last week / (60) last 3 months / (250) last year /(1000) last year
# by finding the ratio between today's closing price and the closing price in those periods to factor in the 
# the condition of the market, e.g. it may be in a downturn/upswing
horizons = [2,5,60,250,1000]

new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50,random_state=1)
```

```python
# editing the 
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)

    return combined
```

```python
predictions = backtest(sp500, model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
